# Exercise
In exercise folder (same level as this notebook on github) there is hiring.csv. This file contains hiring statics for a firm such as experience of candidate, his written test score and personal interview score. Based on these 3 factors, HR will decide the salary. Given this data, you need to build a machine learning model for HR department that can help them decide salaries for future candidates. Using this predict salaries for following candidates,

2 yr experience, 9 test score, 6 interview score

12 yr experience, 10 test score, 10 interview score

#### Answer
53713.86 and 93747.79

```python
url3="https://raw.githubusercontent.com/codebasics/py/master/ML/2_linear_reg_multivariate/Exercise/hiring.csv"
```

```python
df11=pd.read_csv(url3)
df11
```

```python
import word2number.w2n as w
```

```python
df11['experience']=df11['experience'].fillna('zero')
df11
```

```python
df11.experience = df11.experience.apply(w.word_to_num)
df11
```

```python
df11['test_score(out of 10)']=df11['test_score(out of 10)'].fillna(df11['test_score(out of 10)'].median())
df11
```

```python
df13=df11.drop('salary($)',axis='columns')
df13
```

```python
sal=df11['salary($)']
sal
```

```python
model4=lm.LinearRegression()
```

```python
model4.fit(df13,sal)
```

```python
model4.predict([[2,9,6]])
```

```python
model4.score(df13,sal)
```

# Save Model State 
## using Pickle

```python
import pickle
```

```python
with open('model_pickle','wb') as file:
    pickle.dump(model4,file)
```

```python
with open('model_pickle','rb') as file:
    mp = pickle.load(file)
```

```python
mp.predict([[2,9,6]])
```
